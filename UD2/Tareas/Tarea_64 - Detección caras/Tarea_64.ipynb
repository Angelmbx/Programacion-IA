Fai unha aplicación de consola que recolla unha foto e imprima o número de caras que temos na imaxe.

```python
import numpy as np
import cv2 as cv
from matplotlib import pyplot as plt

img_original = cv.imread("minnesota_timberwolves.jpg") 
plt.imshow(img_original)
plt.title('Minnesota Timberwolves 2011 Roster')
plt.xticks([]), plt.yticks([])
```

Fai unha aplicación de consola que recolla unha foto e imprima o número de caras que temos na imaxe.

```python
import numpy as np
import cv2 as cv
from matplotlib import pyplot as plt
```

```python
img = cv.imread("minnesota_timberwolves.jpg") 
img_color = cv.cvtColor(img, cv.COLOR_BGR2RGB)
gray = cv.cvtColor(img_color, cv.COLOR_BGR2GRAY)
plt.imshow(img_color)
plt.title('Minnesota Timberwolves 2011 Roster')
plt.xticks([]), plt.yticks([])
plt.show()

plt.imshow(gray)
plt.title('Imagen en gris')
plt.xticks([]), plt.yticks([])
plt.show()
```

Fai unha aplicación de consola que recolla unha foto e imprima o número de caras que temos na imaxe.

```python
import numpy as np
import cv2 as cv
from matplotlib import pyplot as plt
```

```python
img = cv.imread("minnesota_timberwolves.jpg") 
img_color = cv.cvtColor(img, cv.COLOR_BGR2RGB)
gray = cv.cvtColor(img_color, cv.COLOR_BGR2GRAY)
plt.imshow(img_color)
plt.title('Minnesota Timberwolves 2011 Roster')
plt.xticks([]), plt.yticks([])
plt.show()

plt.imshow(gray)
plt.title('Imagen en gris')
plt.xticks([]), plt.yticks([])
plt.show()
```

```python
# Si no detecta bien las caras en la imagen, retocar los parámetros dentro de 'detectMultiScale'

def buscarCaras(imagen):
  face_cascade = cv.CascadeClassifier("haarcascade_frontalface_default.xml")

  faces = face_cascade.detectMultiScale(
      imagen,
      scaleFactor = 1.1,
      minNeighbors = 22,
      minSize = (30,30), 
      )
  print("Caras = ", len(faces))
  destino = imagen.copy()
  for (x,y,w,h) in faces:
      cv.rectangle(destino, (x,y), (x+h, y+h), (0, 255, 0), 6)

  plt.imshow(destino)
  plt.title('Caras detectadas')
  

  plt.show()
```

```python
wolves = cv.imread("minnesota_timberwolves.jpg")
wolves_color = cv.cvtColor(wolves, cv.COLOR_BGR2RGB)
buscarCaras(wolves_color)
```
